```python
class Student : 
    def __init__(self , name, roll_no, marks):

        self.__name = name 
        self.__roll_no = roll_no
        self.__marks = marks
        self.__grade = self.__calculate_grade()

    def __calculate_grade():

            if self.__marks >= 90:
                return "A+"
            if self.__marks >= 80:
                return "A"
            if self.__marks >= 70:
                return "B"
            if self.__marks >= 60:
                return "C"
            if self.__marks >= 50:
                return "D"
            else:
                return "F"
        
    def get_name(self):
            return self.__name
        
    def get_roll_no(self):
            return self.__roll_no
        
    def get_marks(self):
            return self.__marks
        
    def get_grade(self):
            return self.__grade
        
    def __str__(self):
            return f"Name : {self.__name} , Roll No : {self.__roll_no}, Marks : {self.__marks} , Grade : {self.__grade}"
        

class School : 
    def __init__(self):
        self.__students = []

    def add_student(self , student):
        self.__students.append(student)
        print(f" Student  '{student.get_name()}' added successfully")

    def display_students(self):
        if not self.__students:
                print(f"No Students  enrolled yet")
        else:
            print(" All Students")
            for student in self.__students:
                print(student)

    def search_by_roll_up(self, roll_no):
        for student in self.__students:
            if student.get_roll_no():
                print("Studern Fount :")
                print(student)
                return
            else:
                print("No Student Found")

    def calculate_class_average(self):

        if not self.__students:
            print("No Average to Calculate")
            return
        
        total = sum(student.get_marks() for student in self.__students )
        avg = total / len(self.__students)
        print(f"Average MAtks : {avg : .2f}")


s1 = Student("SEdhu", 1 , 95)
s2 = Student("Kavi", 2 , 60)
s3 = Student("Kesav", 3 , 83)

school = School()

school.add_student(s1)
school.add_student(s2)
school.add_student(s3)


school.display_students()
school.search_by_roll_up(2)
school.calculate_class_average()
```
