```python
import datetime
import json
import os
import uuid

books_file = "books.txt"


def timestamp():
    
    now = datetime.datetime.now()
    return now.strftime("%Y-%m-%d %I:%M:%S %p")


def load_books():
    # Load books 
    if os.path.exists(books_file):
        with open(books_file, "r") as file:
            try:
                return json.load(file)
            except json.JSONDecodeError:
                return []
    return []


def save_books(books):
    with open(books_file, "w") as file:
        json.dump(books, file, indent=4)


def add_book(books):
    name = input("Book Name: ")
    quantity = int(input("Book Quantity: "))
    author = input("Author Name: ")


# Python = python 
    # check if already exists
    for book in books:
        if book["Book Name"].lower() == name.lower() and book["Book Author"].lower() == author.lower():
            book["Book Quantity"] += quantity
            book["Action Performed"] = "Updated"
            book["Timestamp"] = timestamp()
            print(f"Updated quantity for '{name}' by {author} (new quantity: {book['Book Quantity']})")
            return books

    # if not exists, add new
    new_book = {
        "Unique Id": str(uuid.uuid4())[:8],
        "Book Name": name,
        "Book Author": author,
        "Book Quantity": quantity,
        "Action Performed": "Added",
        "Timestamp": timestamp()
    }
    books.append(new_book)
    print(f"Added new book '{name}' by {author} (quantity: {quantity})")
    return books


def update_book(books):
    name = input("Enter Book Name to update: ")
    author = input("Enter Author Name: ")

    for book in books:
        if book["Book Name"].lower() == name.lower() and book["Book Author"].lower() == author.lower():
            new_qty = int(input("Enter new quantity: "))
            book["Book Quantity"] = new_qty
            book["Action Performed"] = "Updated"
            book["Timestamp"] = timestamp()
            print(f"Updated '{name}' by {author} to quantity {new_qty}")
            return books

    print("Book not found!")
    return books


def delete_book(books):
    name = input("Enter Book Name to delete: ")
    author = input("Enter Author Name: ")

    for i, book in enumerate(books):
        if book["Book Name"].lower() == name.lower() and book["Book Author"].lower() == author.lower():
            confirm = input(f"Are you sure you want to delete '{name}' by {author}? (y/n): ").lower()
            if confirm == "y":
                books.pop(i)
                print(f"Book '{name}' by {author} deleted successfully!")
            else:
                print("Deletion cancelled.")
            return books

    print("Book not found!")
    return books


def view_books(books):
    if not books:
        print("No books available.")
        return

    # Header
    print("\nList of Books:")
    print(f"{'ID':<10} {'Book Name':<25} {'Author':<20} {'Qty':<5} {'Action':<10} {'Timestamp'}")
    print("-" * 90)

    # Each book row
    for b in books:
        print(f"{b['Unique Id']:<10} "
              f"{b['Book Name']:<25} "
              f"{b['Book Author']:<20} "
              f"{b['Book Quantity']:<5} "
              f"{b['Action Performed']:<10} "
              f"{b['Timestamp']}")



while True:
    books = load_books() 

    print("\n Book Manager ")
    print("1. Add Book")
    print("2. Update Book")
    print("3. Delete Book")
    print("4. View Books")
    print("5. Save to File")
    print("6. Exit")

    choice = input("Enter choice (1-6): ")

    if choice == "1":
        books = add_book(books)
        save_books(books)
    elif choice == "2":
        books = update_book(books)
        save_books(books)
    elif choice == "3":
        books = delete_book(books)
        save_books(books)
    elif choice == "4":
        view_books(books)
    elif choice == "5":
        save_books(books)
        print("Books saved to books.txt")
    elif choice == "6":
        print("Exiting program.")
        break
    else:
        print("Invalid choice, try again!")
```
